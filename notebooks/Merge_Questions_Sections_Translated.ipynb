```python
import warnings
warnings.filterwarnings("ignore")
import json
import os
from datasets import DatasetDict,Dataset
import pandas as pd
```

```python
versionType  ="v2"          #VQA v2.0 dataset
taskType    ='OpenEnded'    #type of annotations in the JSON file (OpenEnded).
dataType    ='mscoco'       #source of the images (mscoco or abstract_v002).
dataSubType = "train2014"   #type of data subtype (e.g. train2014/val2014/test2015 for mscoco, train2015/val2015 for abstract_v002).

quesFile     = f"{versionType}_{taskType}_{dataType}_{dataSubType}_questions.json"
```

```python
foolder = '/content/drive/MyDrive/questions-sections-translated-dataset'
```



---

# Merge

```python
listpathes = os.listdir(foolder)
lstques = []
for file in listpathes:
    path = os.path.join(foolder,file)
    with open(path, "r") as f:
        questran_dict = json.load(f)

    lstques.extend(questran_dict['questions'])

mergfile = os.path.join(foolder,f'translated_{quesFile}')
questran_dict['questions'] = lstques

with open(mergfile, 'w') as f:
  json.dump(questran_dict, f)
```

```python
questran_dict = json.load(open(mergfile,'r'))
questran_dict.keys()
```
